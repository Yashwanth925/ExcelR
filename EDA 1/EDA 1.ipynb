```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Cardiotocographic.csv')
df
```

```python
df.describe()
```

```python
print(df.shape)
df.isnull().sum()
```

```python
df['LB']=df['LB'].fillna(133.343)
df['AC']=df['AC'].fillna(0.003)
df['NSP']=df['NSP'].fillna(1.304)
df['Tendency']=df['Tendency'].fillna(0.316)
df['Width']=df['Width'].fillna(70.42)
df['MLTV']=df['MLTV'].fillna(8.284)
df['DS']=df['DS'].fillna(0.0)
df['DP']=df['DP'].fillna(0.0)
```

```python
df.isnull().sum()
```

```python
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace=True) ## removes any duplicates present 
df.duplicated().any()
```

```python
corr= df.corr()
corr
```

# Visualizations

```python
# Histograms for numerical variables
df.hist(figsize=(15, 10))
plt.suptitle('Histograms of Numerical Variables')
plt.show()
```

```python
#Correlation Heatmap
plt.figure(figsize= (10,8))
sns.heatmap(corr, cmap='viridis',vmin=-1, vmax=1, center=0,annot=True)
plt.show()  
```

```python
plt.boxplot(df)
plt.title("Detecting the outliners using boxplot")
plt.show()
```

```python
Q3= df.quantile(0.75)
Q1= df.quantile(0.25)
IQR= Q3-Q1
lower_bound= Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df_cleaned = df[~((df < lower_bound) | (df > upper_bound)).any(axis=1)]
df_cleaned
```

```python
# Visualize the data before and after removing outliers
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.boxplot(df.values)
plt.title('Before Removing Outliers')
plt.xticks(range(1, len(df.columns) + 1), df.columns, rotation=45)

plt.subplot(1, 2, 2)
plt.boxplot(df_cleaned.values)
plt.title('After Removing Outliers')
plt.xticks(range(1, len(df_cleaned.columns) + 1), df_cleaned.columns, rotation=45)

# plt.tight_layout()
plt.show()
```

##### Identified potential outliers in several features using Boxplot, filling missing values with the mean of each column and removed outliers using the Interquartile Range (IQR) method.

##### The second subplot will also have missing values filled with the mean and outliers removed.
